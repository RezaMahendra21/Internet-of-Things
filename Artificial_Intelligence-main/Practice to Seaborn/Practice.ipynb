# 1. Use on of the dataframes provided by Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np

import warnings; warnings.simplefilter('ignore')

taxis = sns.load_dataset("taxis")
taxis.head()
```

```python
taxis.info()
```

```python
taxis.describe()
```

```python
taxis = taxis[taxis['passengers'] != 0]
```

```python
taxis.describe()
```

# 2. Apply basic charts
- Bar and scatter charts

```python
sns.barplot(data = taxis, x = 'passengers', y = 'tip', errorbar = None, hue = 'passengers')
```

```python
sns.scatterplot(data = taxis, x = 'distance', y = 'total')
```

# 3. Apply statistical charts
- Boxplot, violin and joint plots

```python
sns.boxplot(data = taxis, x = 'pickup_borough', y = 'total')
```

```python
sns.violinplot(data = taxis, x = 'dropoff_borough', y = 'total')
```

```python
sns.jointplot(data = taxis, x = 'distance', y = 'total', hue = 'passengers', kind = 'kde')
```

# 4. Study at least two more by yourself and apply it to the dataset

```python
sns.displot(data = taxis, x = 'tip', kind = 'kde', hue = 'passengers', multiple = 'stack')
```

```python
taxis_new = taxis.select_dtypes(include=['int64', 'float64'])
taxis_corr = taxis_new.corr()
sns.heatmap(data = taxis_corr)
```
