```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.countplot(x='sex', data=titanic)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")

sns.countplot(x='sex', 
    data=titanic, 
    hue='embark_town', 
    palette='Pastel2')
plt.show()
```

```python
titanic = sns.load_dataset("titanic")

sns.countplot(x="pclass", data=titanic, hue="survived", palette="husl")
plt.title("Survived by Passanger Class")
plt.xlabel("Passenger Classes")
plt.ylabel("Number of Passengers")
plt.legend(loc='upper left')
plt.show()
```

```python
titanic = sns.load_dataset("titanic")

ax = sns.barplot(x='deck', y='fare', hue='deck', data=titanic, palette="pastel", errorbar=None, legend=False)
plt.title("Fare prices by Deck")
plt.xlabel("deck")
plt.ylabel("fare")

for p in ax.containers:
    ax.bar_label(p, fmt='%.2f')
    
plt.show()
```

```python
sns.histplot(x='age', data = titanic)
```

```python
sns.histplot(x='age', data = titanic, hue='sex', kde = True)
plt.title("Number of Passengers by Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
sns.histplot(x='age', 
    data = titanic, 
    hue='sex', 
    element = 'step')
```

```python
sns.displot(titanic, x='age', kind='ecdf', hue='pclass')

plt.title("Distribution of Passengers by Age")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
sns.displot(titanic, x='age', kind='kde', hue='pclass')

g = sns.FacetGrid(titanic, 
    col="pclass",  
    row="survived", 
    hue="sex")
g.map(sns.kdeplot, "age", shade = True)
```
