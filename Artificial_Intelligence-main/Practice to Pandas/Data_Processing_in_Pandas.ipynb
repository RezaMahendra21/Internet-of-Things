```python
technologies = {
    'Cources': ['Spark', 'PySpark', 'Hadoop', 'Python', 'pandas'],
    'Fee': [20000, 25000, 26000, 22000, 24000],
    'Duration': ['30days', '40days', '35days', '40days', '60days'],
    'Discount': [1000, 2300, 1200, 2500, 2000]
}

index_labels = ['r1', 'r2', 'r3', 'r4', 'r5']

df = pd.DataFrame(technologies, index = index_labels)
print(df)
```

```python
print(df.loc['r2'])

print('-' * 20)

print(df.iloc[1])
```

```python
print(df.loc[:, 'Cources'])

print('-' * 20)

print(df.iloc[:, 0])
```

```python
print(df.loc[['r2', 'r3']])

print('-' * 20)

print(df.iloc[[1, 2]])
```

```python
print(df.loc[:, ['Cources', 'Fee', 'Discount']])

print('-' * 20)

print(df.iloc[:, [0, 1, 3]])
```
