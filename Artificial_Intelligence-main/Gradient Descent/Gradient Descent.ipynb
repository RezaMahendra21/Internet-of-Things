```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def grad_f(x):
    return 2*x

def gradient_descent(starting_point, learning_rate, n_iterations):
    x = starting_point
    history = [x]
    for _ in range(n_iterations):
        gradient = grad_f(x)
        x = x - learning_rate * gradient
        history.append(x)
    return np.array(history)

starting_point = 10
learning_rate = 0.1
n_iterations = 50

history = gradient_descent(starting_point, learning_rate, n_iterations)

x = np.linspace(-10, 10, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^2')
plt.scatter(history, f(history), color='red')
plt.plot(history, f(history), color='red', linestyle='dotted')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
learning_rates = [0.01, 0.1, 0.5, 1.0]
colors = ['blue', 'green', 'orange', 'purple']
starting_point = 10

plt.figure(figsize=(12, 8))
plt.plot(x, y, label='f(x) = x^2')

for lr, color in zip(learning_rates, colors):
    history = gradient_descent(starting_point, lr, n_iterations)
    plt.scatter(history, f(history), color=color)
    plt.plot(history, f(history), color=color, linestyle='dotted', label=f'LR = {lr}')

plt.title('Impact of Learning Rate on Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
