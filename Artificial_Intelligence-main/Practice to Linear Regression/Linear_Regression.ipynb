```python
x = diabetes.data[:, 2]
y = diabetes.target
```

```python
w = 1.0 # gradient
b = 1.0 # bias
```

```python
y_hat = x[0] * w + b
print(y_hat)
```

```python
print(y[0])
```

```python
i = 0

# first trial
# learning_rate = 0.1 # 0.1, 1, 10, 100, 1000, 10000 등 다양하게 변화시키면서 y, y_hat_inc의 수치 변화 확인

# second trial
# learning_rate = 1

# third trial
# learning_rate = 10

# forth trial
# learning_rate = 100

# fifth trial
# learning_rate = 1000

# sixth trial
learning_rate = 200

while (i < 10):
    w_inc = w + learning_rate
    y_hat_inc = x[0] * w_inc + b
    print(f'updated y_hat_inc = {y_hat_inc}. error: y-y_hat_inc = {y[0] - y_hat_inc}.')
    learning_rate += 200
    i += 1
```

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)

b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```python
w = 1.0
b = 1.0

w_rate =(y_hat_inc - y_hat) / (w_inc - w)
w_inc = w + w_rate
print(w_rate)
y_hat_inc = x[0] * w_inc + b
print(f'updated y_hat_inc = {y_hat_inc}, y - y_hat_inc = {y[0] - y_hat_inc}')
```

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```

```python
w_new = w + w_rate
print(w_new)
```

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)

b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```
