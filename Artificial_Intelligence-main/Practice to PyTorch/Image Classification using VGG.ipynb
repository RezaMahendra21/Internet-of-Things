# 2. Load VGG-16 Model

# 3. Data Preprocessing

```python
class BaseTransform():
    def __init__(self, resize, mean, std):
        self.base_transform = transforms.Compose([
            transforms.Resize(resize),
            transforms.CenterCrop(resize),
            transforms.ToTensor(),
            transforms.Normalize(mean, std)
        ])
        
    def __call__(self, img):
        return self.base_transform(img)
```

```python
# Check the image preprocessing

# 1. Load a image
image_file_path = './pytorch_advanced/1_image_classification/data/goldenretriever-3724972_640.jpg'
img = Image.open(image_file_path) # [height][width][RGB]

# 2. Show the image
plt.imshow(img)
plt.show()

# 3. image preprocessing and show the result
resize = 224
mean = (0.485, 0.456, 0.406)
std = (0.229, 0.224, 0.225)
transform = BaseTransform(resize, mean, std)
img_transformed = transform(img)

# Transformation to (RGB, height, width) -> (height, width, RGB) and the range of 0~1
img_transformed = img_transformed.numpy().transpose((1, 2, 0))
img_transformed = np.clip(img_transformed, 0, 1)
plt.imshow(img_transformed)
plt.show()
```

# 4. Create Postprocessing Class

```python
ILSVRC_class_index = json.load(open('./pytorch_advanced/1_image_classification/data/imagenet_class_index.json', 'r'))
ILSVRC_class_index
```

```python
class ILSVRCPredictor():
    def __init__(self, class_index):
        self.class_index = class_index
        
    def predict_max(self, out):
        maxid = np.argmax(out.detach().numpy())
        predicted_label_name = self.class_index[str(maxid)][1]
        
        return predicted_label_name
```

# 5. Prediction

```python
ILSVRC_class_index = json.load(open('./pytorch_advanced/1_image_classification/data/imagenet_class_index.json', 'r'))
predictor = ILSVRCPredictor(ILSVRC_class_index)

image_file_path = './pytorch_advanced/1_image_classification/data/goldenretriever-3724972_640.jpg'
img = Image.open(image_file_path)

transform = BaseTransform(resize, mean, std)
img_transformed = transform(img)
inputs = img_transformed.unsqueeze_(0)

out = net(inputs)
result = predictor.predict_max(out)

print(f'Prediction: {result}')
```
